```python
import fv3fit
import datetime
import xarray as xr
import intake
import cftime
import os
import numpy as np
import time
import matplotlib.pyplot as plt
import pandas as pd
import vcm
from vcm.catalog import catalog

grid = catalog['grid/c48'].read()

n_weeks=16
time_sample_freq="12H"

def save_fig(name: str, dpi: int=150,):
    plt.savefig(f"figures/pdf/{name}.pdf", dpi=dpi, bbox_inches='tight')
    plt.savefig(f"figures/png/{name}.png", dpi=dpi, bbox_inches='tight')
    plt.savefig(f"figures/eps/{name}.eps", format="eps", bbox_inches='tight')
    
time_tick_loc = [cftime.DatetimeJulian(2020, 2, 1), cftime.DatetimeJulian(2020, 3, 1), cftime.DatetimeJulian(2020, 4, 1), cftime.DatetimeJulian(2020, 5, 1)]

baseline_hash = "61c5f8380a6560540d65ba0b9b9fb41c"
ml_tquv_seed_0_hash = "286d3f0e7fcb9b8ce3f8d3a61299c85f"
ml_tquv_seed_3_hash = "9b944bf2627c59de29d2ef5762825834"
```

```python
gamma_4_79_model_url = 'gs://vcm-ml-experiments/claytons/2022-08-09/tq-combined-nd-trial-1/trained_models/ocsvm-larger-gamma'

diags_ = []
for run_hash in [baseline_hash, ml_tquv_seed_0_hash , ml_tquv_seed_3_hash]:
    diags_url_ = os.path.join(gamma_4_79_model_url, "diags_novelty_offline", run_hash)
    diags_.append(intake.open_zarr(diags_url_).read())
    
diags = xr.concat(diags_, dim=pd.Index(["Baseline", "$g_\mathrm{Tquv}$ seed 0 \n (higher PWAT error)", "$g_\mathrm{Tquv}$ seed 3 \n (lower PWAT error)"], name="run"))

zonal_avg_diags = vcm.zonal_average_approximate(
    grid.lat, diags, 
)

tick_loc = np.array([-60, -30, -15, 0, 15, 30, 60])
tick_loc_sin = np.sin(tick_loc * np.pi/180)
sin_lat= np.sin(zonal_avg_diags.lat.values * (np.pi/180))

zonal_avg_diags  = zonal_avg_diags.assign_coords(lat=sin_lat)['is_novelty']
```

```python

zonal_avg_diags = zonal_avg_diags.assign_coords(
    {"run": ["Baseline", "$g_\mathrm{Tquv}$ seed 0 \n (higher PWAT error)", "$g_\mathrm{Tquv}$ seed 3 \n (lower PWAT error)"]}
)

f = zonal_avg_diags.plot(
    x='time',
    col='run', 
    figsize=(10,3),
    rasterized=True
)
for i, subtitle in enumerate(zonal_avg_diags.run.values):
    ax = f.axes[0][i]
    plt.setp(ax.xaxis.get_majorticklabels(), rotation=50)
    ax.set_title(subtitle)


f.axes[0][0].set_ylabel("Latitude [deg]")    

#plt.suptitle("ML-corrected $g_{Tquv}$", x=0.425, y=1.05, fontsize=16)
plt.xticks(time_tick_loc)
plt.yticks(tick_loc_sin, tick_loc)  # Set label locations.
f.cbar.ax.set_ylabel("zonal-mean novelty fraction")

#save_fig('offline_novelty_fraction_gamma_4_79')
```
