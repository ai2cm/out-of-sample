```python
import matplotlib.pyplot as plt
from PIL import Image
import math
import os
import subprocess
import proplot


DPI=300
fig_dir="figures/final"

def save_figs(fig_name: str, fig_dir=fig_dir):
    pdf_file = os.path.join(fig_dir, "pdf", f"{fig_name}.pdf")
    temp_eps_file = os.path.join(fig_dir, "pdf", f"{fig_name}.eps")
    eps_file = os.path.join(fig_dir, "eps", f"{fig_name}.eps")
    plt.savefig(pdf_file, dpi=DPI, bbox_inches="tight")
    subprocess.check_call(["pdftops", "-eps", pdf_file])
    subprocess.check_call(["mv", temp_eps_file, eps_file])


def open_eps(filename, dpi=DPI):
    img = Image.open(filename)
    original = [float(d) for d in img.size]
    scale = dpi/72.0            
    if dpi is not 0:
        img.load(scale = math.ceil(scale))
    if scale != 1:
        img.thumbnail([round(scale * d) for d in original], Image.ANTIALIAS)
    return img
```

#### RMSE scatter plots

```python

a = open_eps('figures/eps/rmse_scatter_precip.eps', dpi=DPI)
b = open_eps('figures/eps/rmse_scatter_t850.eps', dpi=DPI)

c = open_eps('figures/eps/rmse_scatter_pwat.eps', dpi=DPI)
d = open_eps('figures/eps/rmse_scatter_t200.eps', dpi=DPI)
```

```python

f, axes = plt.subplots(2, 2, figsize=(5,4))

fsize=10
letter_y = 0.97

ax=axes[0][0]
ax.imshow(a)
ax.axis("off")
ax.text(0, letter_y, "a)", va='top', fontsize=fsize, transform=ax.transAxes)

ax=axes[0][1]
ax.imshow(b)
ax.axis("off")
ax.text(0, letter_y, "b)", va='top', fontsize=fsize, transform=ax.transAxes)

ax=axes[1][0]
ax.imshow(c)
ax.axis("off")
ax.text(0, letter_y, "c)", va='top', fontsize=fsize, transform=ax.transAxes)

ax=axes[1][1]
ax.imshow(d)
ax.axis("off")
ax.text(0, letter_y, "d)", va='top', fontsize=fsize, transform=ax.transAxes)

plt.tight_layout()

save_figs(fig_name="rmse_scatter")
```
